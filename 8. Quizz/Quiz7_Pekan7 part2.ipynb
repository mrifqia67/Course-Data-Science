```python
# melakukan pembersihan 
#1. Case folding
#2. Stopwords

import pandas as pd
import re
import string

from nltk.corpus import stopwords
```

```python
# memanggil data
df = pd.read_csv('kompas_20211201.csv')
display(df)
```

```python
#1. Case folding
# membuat kumpulan kata pada kolom tersebut menjadi lowercase   
df['title'] = df['title'].str.lower()

# menghilangkan alamat url
df['title'] = df['title'].replace('http\S+|www\S+|https\S+', '', regex = True)  

# mendelete username
df['title'] = df['title'].replace('@[\w]+', '', regex = True)

# menghapus angka yang terdapat dalam setiap tweet
df['title'] = df['title'].replace('\d+', '', regex = True)

# menghapus tanda baca
df['title'] = df['title'].str.translate(str.maketrans("","",string.punctuation))

# menghapus whitespace
df['title'] = df['title'].str.strip()

# menghilangkan karakter khusus
df['title'] = df['title'].apply(lambda x: x.encode('ascii', 'ignore').decode('ascii'))

# cek hasil
for i in df['title']:
    print(i)
```

```python
#2. stopwords
stopwords_indo = stopwords.words('indonesian')

df['title'] = df['title'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords_indo)]))

# cek hasil
for i in df['title']:
    print(i)
```

```python
# simpan file dengan nama kompas_20211201_clean.csv
df.to_csv('kompas_20211201_clean.csv', index = False, header=True)
```
