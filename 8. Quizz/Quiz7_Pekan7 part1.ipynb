```python
# memanggil library
import pandas as pd
import csv

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# page 1
website = 'https://indeks.kompas.com/?site=all&date=2021-12-01'
html = urlopen(website)
data = BeautifulSoup(html, 'html.parser')

test = data.findAll('div', {'class':'latest--indeks'})[0]
rows = test.findAll('div')

# mencari submenu pada inspect
hasil1 = []
for row in rows:
    for cell in row.findAll(['h3']):
        hasil1.append(cell.get_text())
```

```python
# page 2
website = 'https://indeks.kompas.com/?site=all&date=2021-12-01&page=2'
html = urlopen(website)
data = BeautifulSoup(html, 'html.parser')    
    
test = data.findAll('div', {'class':'latest--indeks'})[0]
rows = test.findAll('div')    
 
# mencari submenu pada inspect
hasil2 = []
for row in rows:
    for cell in row.findAll(['h3']):
        hasil2.append(cell.get_text())    
```

```python
# menggabungkan data dari page 1-5
data = hasil1 + hasil2 + hasil3 + hasil4 + hasil5
data
```

```python
# cleaning data karena ada data duplikat
clean = []
for i in data:
    if i not in clean:
        clean.append(i)       
        
# data = 2021 - 12 - 01
# membuat dataframe
df = pd.DataFrame({'date' : '2021-12-01',
                   'title' : clean})
display(df)
```

```python
# menyimpan dalam bentuk csv dengan nama kompas_20211201.csv
df.to_csv('kompas_20211201.csv', index = False, header=True)
```
