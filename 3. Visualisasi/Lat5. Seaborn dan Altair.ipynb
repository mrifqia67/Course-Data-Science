```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# import library seaborn
import seaborn as sns
```

```python
df = sns.load_dataset('diamonds')
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
# melihat isi data
df.head(7)
```

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"
 
# code here
# plotting the data
sns.scatterplot(data = df, x='carat', y='price', hue = 'clarity', size = 'depth',
                sizes=(2, 10), alpha=1, linewidth = 0, palette = palette)
sns.set()
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here
with sns.axes_style(style='ticks'):
    sns.catplot(data=df, x='clarity', y='price', col = 'cut', kind = 'box', sym = '')
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
# code here
# mengambil 300 data secara acak dari kolom carat, depth, table , dan price
data = df.loc[:,['cut','carat','depth','table','price']]
data = data.sample(n=300, random_state=123)

# plotting the data
sns.pairplot(data, hue='cut', markers = ['p', 'o', 's', 'D', 'X'], corner = True)
sns.set()
```

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

```python
# menggunakan data nomor 3
# code here
with sns.axes_style('white'):    
    # expected output 1
    sns.jointplot(x = data['carat'], y = data['price'], kind = 'kde')
    
# code    
with plt.style.context('seaborn'):
    # expected output 2
    sns.jointplot(x = data['carat'], y = data['price'], kind = 'reg')
```
