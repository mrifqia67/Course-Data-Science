<h2>Soal 1:</h2>

Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# membuat plot biasa
# membuat component figure dan axis
figur, axis = plt.subplots()

# memberikan data kedalam axis
axis.plot(x, cos)
axis.plot(x, cos3)
axis.plot(x, sin)
axis.plot(x, sin2)

plt.show()
```

```python
# membuat custom plot
# membuat component figure dan axis dan memberi ukuran figurnya
figure, axis1 = plt.subplots(figsize=(10,6.5))

# memberikan data kedalam axis dan memberi legendnya 
axis1.plot(x, cos, linewidth=1.3, color = 'r', label = 'Cos x') #grafik 3
axis1.plot(x, cos3, linewidth=1.3, linestyle='--', color = 'g', label = '3 Cos x') #grafik 4
axis1.plot(x, sin, linewidth=1.3, linestyle='-.', color = 'b', label = 'Sin(x-90)') #grafik 2
axis1.plot(x, sin2, linewidth=1.3, marker = 'x', linestyle='none', color = 'y', label = '2 sin x') #grafik 1

# memberi judul pada sumbu x,y dan judul grafik
axis1.set_xlabel('X Axis Label',fontsize = 10)
axis1.set_ylabel('Y Axis Label',fontsize = 10)
axis1.set_title('Grafik Cos Sin',fontsize = 10)

# meletakkan legend di tengah figure dan adjust the legend
plt.legend(loc = 10,bbox_to_anchor=(0.47,0.475),framealpha=0.4,fontsize=8)
plt.show()
```
