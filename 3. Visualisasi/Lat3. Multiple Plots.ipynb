```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
# membuat kertas gambar terlebih dahulu
figur, akses = plt.subplots(nrows=1, ncols=3, figsize=(24,10), sharey = True)

# plot masing-masing 
b = akses[0].plot(x,tan, color = 'b')
r = akses[1].plot(x,sin, color = 'r')
g = akses[2].plot(x,cos, color = 'g')

# memberi judul untuk masing-masing grafik
akses[0].set_title('Grafik tanx',fontsize = 14)
akses[1].set_title('Grafik sinx',fontsize = 14)
akses[2].set_title('Grafik cosx',fontsize = 14)

# memberi legend
figur.legend(handles = [b[0],r[0],g[0]], labels = ['tanX','sinX','cosX'], loc = 'upper center', fontsize  = 12)

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
# membuat figur utama
fig, ax = plt.subplots(figsize=(13,8))
ax.plot(x, y)
ax.set_title('Grafik CosX',fontsize = 10)

# membuat anak figur
ins1 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins1.plot(x, y2, color = 'k')
ins1.plot(x, y, color = 'b')
ins1.set_title('Grafik CosX^4',fontsize = 10)

ins2 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins2.plot(x, y3, color = 'y')
ins2.plot(x, y, color = 'b')
ins2.set_title('Grafik CosX^5',fontsize = 10)

ins3 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins3.plot(x, y4, color = 'r')
ins3.plot(x, y, color = 'b')
ins3.set_title('Grafik CosX^2',fontsize = 10)

ins4 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins4.plot(x, y5, color = 'g')
ins4.plot(x, y, color = 'b')
ins4.set_title('Grafik CosX^3',fontsize = 10)

plt.show()
```
