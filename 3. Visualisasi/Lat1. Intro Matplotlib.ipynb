**Figure** adalah halaman dalam objek visual dan bersifat polos, serperti kanvas atau kertas gambar.

**Axis** adalah suatu area di dalam figure dimana data akan di plot. Axis terdiri atas suatu koordinat x dan y dalam memetakan suatu data.

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# figure dan axis kosong tanpa data
kosong1 = plt.figure()
data = kosong1.add_subplot()
```

```python
# Plot data dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit
# secara explisit
x = np.linspace(0, 20, 100)
y = np.cos(x/2)

kosong1 = plt.figure()
data = kosong1.add_subplot()

data.plot(x, y)
plt.show()
```

```python
# secara implisit
plt.plot(x, y)
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
# Plot ketiga data berikut ke dalam satu axis menggunakan seaborn style

with plt.style.context('seaborn'):
    fig, axis = plt.subplots()
    axis.plot(x,y)
    axis.plot(x,y2)
    axis.plot(x,y3)
    plt.show()
```

```python
# save data tersebut dan menampilkan kembali gambar tersebut
fig.savefig('axis1.png')

# menampilkan figure yang disimpan
from IPython.display import Image
display(Image('axis1.png'))
```
