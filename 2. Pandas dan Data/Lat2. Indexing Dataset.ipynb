```python
# Persiapan membuat data
import pandas as pd
import random
import numpy as np

pelajaran = ['Matematika', 'Fisika', 'Biologi', 'Kimia', 'Geologi', 'Ekonomi']
kelas = ['12A', '12B', '12C', '12D']

# menggunakan random.choice untuk memilih nilai untuk variabel
hasil_nilai = {
    'siswa_id' : range(1, 21),
    'pelajaran' : [random.choice(pelajaran) for i in range(20)],
    'nilai' : [random.choice(range(0, 100)) for i in range(20)],
    'jenis-kelamin' : [random.choice(['laki-laki', 'perempuan']) for i in range(20)],
    'kelas' : [random.choice(kelas) for i in range(20)]
}

df = pd.DataFrame(hasil_nilai)
df.head()
```

<h2>Soal 1: Indexing</h2>

```python
# code here
# 1. mengakses kolom nilai dan kelas saja
df.loc[:,['nilai','kelas']]
```

```python
# 2. mengakses row 5 sampai 10 beserta column pelajaran dan nilai
df.loc[5:10:,['pelajaran','nilai']]
```

<h2>Soal 2: Filter DataFrame Dengan Boolean</h2>

1. Tampilkan semua data siswa pada mata pelajaran Ekonomi dan Fisika
2. Tampilkan semua data siswa yang mempunyai nilai lebih dari 50 di mata pelajaran Kimia 
3. Tampilkan semua data yang mempunyai nilai lebih dari 70, berjnis kelamin perempuan dan di kelas 12C

```python
# code here
# 1. menampilkan semua data siswa pada mata pelajaran Ekonomi dan Fisika
mapel = (df.pelajaran == 'Ekonomi') | (df.pelajaran == 'Fisika')
df[mapel]
```

```python
# 2. menampilkan semua data siswa yang mempunyai nilai lebih dari 50 di mata pelajaran Kimia
nilai_kimia = (df.pelajaran == 'Kimia') & (df.nilai >= 50)
df[nilai_kimia]
```

```python
# 3. menampilkan semua data yang mempunyai nilai lebih dari 70, berjnis kelamin perempuan dan di kelas 12C
data = (df.nilai >= 70) & (df['jenis-kelamin'] == 'perempuan') & (df.kelas == '12C')
df[data]
```

<h2>Soal 3: Data Transformation</h2>

lakukan data transformasi terhadap column 'nilai' dengan fungsi transformasi, berikut rumus dari normalisasi

![alt text](https://drive.google.com/uc?id=1wed-EbRSTA0JP1ceDp6lcKBqH2yJ4aa3)

```python
# cara_1
hasil = []
for i in df['nilai']:
    trans =  (i - df['nilai'].min()) / (df['nilai'].max() - df['nilai'].min())
    hasil.append(trans)
    
bulat = [round(num, 2) for num in hasil]
bulat
```

```python
# cara_2
# code here
# data transformation terhadap kolom nilai

# nilai minimum dan maximum
print(df['nilai'].min())
print(df['nilai'].max())

# transformasi
def trans(data):
    return (data - data.min()) / (data.max() - data.min())

hasil = df[['nilai']].apply(trans)
hasil
```
